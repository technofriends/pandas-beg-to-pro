```python
# load the dataframe
import pandas as pd

df = pd.read_csv('airbnb.csv', index_col="id", low_memory=False)
```

```python
df.info()
```

```python
# lets convert the price column to float
df['price'] = df['price'].str.replace('$', '')
df['price'] = df['price'].str.replace(',', '')
df['price'] = df['price'].astype(float)
```

```python
# sort basis the price

df[['NAME','host id','price','country']].sort_values(by='price', ascending=False).head(10)
```

```python
df[['NAME','host id','price','service fee','country']].sort_values(by='price', ascending=True).head(10)
```

```python
columns_to_sort_on = ['price', 'minimum nights']
df[['NAME','host id','price','service fee','country','minimum nights']].sort_values(by=columns_to_sort_on, ascending=False).head(10)
```

```python
select_columns = ['NAME','host id','price','service fee','country','minimum nights']
columns_to_sort_on = ['price', 'minimum nights']

df[select_columns].sort_values(by=columns_to_sort_on, ascending=[False, True]).head(10)
```

```python
# convert service fee to float
df['service fee'] = df['service fee'].str.replace('$', '')
df['service fee'] = df['service fee'].str.replace(',', '')
df['service fee'] = df['service fee'].astype(float)
```

```python
price_less_than_200 = df["price"] < 200
minimum_nights_greater_than_2 = df["minimum nights"] > 2
service_fee_less_than_50 = df["service fee"] < 30

df1 = df[price_less_than_200 & minimum_nights_greater_than_2 & service_fee_less_than_50][my_vars].sort_values(by=columns_to_sort_on, ascending=[False, True])
```
